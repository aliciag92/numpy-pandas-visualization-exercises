```python
#define figure
plt.figure(figsize = (10, 10))

#plot data
x1 = list(range(0, 8))
y1 = [0 * x + 1 for x in x1]

x2 = list(range(1, 8))
y2 = [math.log(x) for x in x2]

x3 = list(range(0, 8))
y3 = [x for x in x3] 

x4 = list(range(1, 8))
y4 = [x*math.log(x) for x in x4] 

x5 = list(range(0, 6))
y5 = [x**2 for x in x5]

x6 = list(range(0, 5))
y6 = [2**x for x in x6]

x7 = list(range(1, 5))
y7 = [math.factorial(x) for x in x7]

x8 = list(range(0, 2))   
y8 = [x**x for x in x8]

plt.plot(x1, y1, label = "$O(1)$") #blue
plt.plot(x2, y2, label = "$O(log n)$") #orange
plt.plot(x3, y3, label = "$O(n)$") #green
plt.plot(x4, y4, label = "$O(n log n)$") #red
plt.plot(x5, y5, label = "$O(n^2)$") #purple
plt.plot(x6, y6, label = "$O(2^n)$") #brown
plt.plot(x7, y7, label = "$O(n!)$") #pink
plt.plot(x8, y8, label = "$O(n^n)$") #black

#customize
plt.title("Big O Notation", fontsize = 12)
plt.xlabel("Elements", fontsize = 12)
plt.ylabel("Operations", fontsize = 12)

#show
plt.legend()
plt.show()
```
